### Importing the Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression

import warnings
warnings.filterwarnings("ignore")
```

### Loading the Dataset

```python
df = pd.read_csv("winequalityN.csv")
print(df.head())
```

```python
df.info()
```

### Describing the data for checking

```python
df.describe().T
```

### Finding the amount of null values to either remove or fill them

```python
df.isnull().sum()
```

### Filling the null value fields with the mean value of the column

```python
for cols in df.columns:
    if df[cols].isnull().sum() > 0:
        df[cols] = df[cols].fillna(df[cols].mean())
df.isnull().sum()
```

### Drawing a histogram to visualize the data

```python
df.hist(bins=20, figsize = (20,15))
plt.show()
```

### Plotting a bar graph to see the number of wines in each quality

```python
plt.bar(df['quality'], df['alcohol'])
plt.xlabel('quality')
plt.ylabel('alcohol')
plt.show()
```

### Drawing a heatmap to reduce the data redundancy

```python
for col in df.columns:
    if df[col].dtype == 'object':
        try:
            df[col] = pd.to_numeric(df[col], errors='coerce')  # Convert to numeric, replace non-convertibles with NaN
        except:
            pass

plt.figure(figsize=(12, 12))
sb.heatmap(df.corr() > 0.7, annot=True, cbar=False)
plt.show()
```

### From the above heatmap we can see that the "Free Sulphur Dioxide" and the "Total sulfur dioxide" are very correlated, now we remove 1 of them as they are being redundant

```python
df = df.drop('total sulfur dioxide', axis = 1)
```

```python
df.head()
```

### Model development

```python
df['best quality'] = [1 if x > 5 else 0 for x in df.quality]
```

```python
df.replace({'white': 1, 'red': 0}, inplace=True)
```

```python
#features = features.fillna(features.mean())
features = df.drop(['quality', 'best quality'], axis=1)
target = df['best quality']

xtrain, xtest, ytrain, ytest = train_test_split(
    features, target, test_size=0.2, random_state=40)

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')  
xtrain = imputer.fit_transform(xtrain)
xtest = imputer.transform(xtest)

xtrain.shape, xtest.shape
```

### Now Normalize te data

```python
n = MinMaxScaler()
xtrain = n.fit_transform(xtrain)
xtest = n.transform(xtest)
```

### Now check for the accuracy for all the 3 models

```python
models = [LogisticRegression(), XGBClassifier(), SVC(kernel='rbf')]

for i in range(3):
    models[i].fit(xtrain, ytrain)

    print(f'{models[i]} : ')
    print('Training Accuracy : ', metrics.roc_auc_score(ytrain, models[i].predict(xtrain)))
    print('Validation Accuracy : ', metrics.roc_auc_score(
        ytest, models[i].predict(xtest)))
    print()
```
